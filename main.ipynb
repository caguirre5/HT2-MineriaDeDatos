<h2>Limpieza de datos<h2>

De acuerdo con la informacion brindad por la Universidad de Wisconsin, sobre estos datos, podemos observar que hay 16 registros que tienen un valor "?" en la columna "Bare Nuclei", por lo cual antes de decidir que tipo de limpieza hacer con los mismos, hacemos un analisis de los mismos.

```python
print(len(df[df['Bare Nuclei'] == '?']))
print(len(df[(df['Class'] == 2) & (df['Bare Nuclei'] == '?')]))
print(len(df[(df['Class'] == 4) & (df['Bare Nuclei'] == '?')]))
```

Como podemos observar de los 16 registros con un valor ? (desconocido), 14 son benignos y 2 son malignos, por lo cual no sería una buena opcion descartarlos, pues no estarían balanceados, por lo cual se decidió reemplazarlos con la mediana de la columna, filtrando los datos por la columna class.


```python
#Filtramos los registros con Class == 2 (benigno)
class_2_df = df[df['Class'] == 2]
#Sacamos la mediana de dichos registros
median = class_2_df['Bare Nuclei'].median()
#Reemplazamos los valores nulos de los registros con class == 2 por la mediana obtenida
df.loc[df['Class'] == 2, 'Bare Nuclei'] = df.loc[df['Class'] == 2, 'Bare Nuclei'].fillna(median)
```

```python
#Filtramos los registros con Class == 4 (maligno)
class_2_df = df[df['Class'] == 4]
#Sacamos la mediana de dichos registros
median = class_2_df['Bare Nuclei'].median()
#Reemplazamos los valores nulos de los registros con class == 4 por la mediana obtenida
df.loc[df['Class'] == 4, 'Bare Nuclei'] = df.loc[df['Class'] == 4, 'Bare Nuclei'].fillna(median)
```

```python
#Ahora que todos los valores son numericos, podemos hacer un cast
df['Bare Nuclei'] = df['Bare Nuclei'].astype('int64')
```

Eliminamos la columna con el codigo de muestra, pues no tiene relevancia en nuestro modelo.

```python
df = df.drop('Sample code number', axis=1)
```

Reemplazar la columna Bare Nuclei a una columna categórica binaria

```python
df['Class'] = df['Class'].replace({2: 0, 4: 1})
```

<h2>Modelo de Regresión Logistica<h2>

Partir en Train y Test

```python
from sklearn.model_selection import train_test_split
X_entreno, X_prueba, y_entreno, y_prueba = train_test_split(df.drop('Class' , axis = 1), 
                                                    df['Class'], test_size = 0.30, 
                                                    random_state = 101)
```

Entrenamiento y prediccion

```python
from sklearn.linear_model import LogisticRegression

modeloLog = LogisticRegression(max_iter = 500)
modeloLog.fit(X_entreno,y_entreno)

y_pred = modeloLog.predict(X_prueba)
y_pred
```

Evaluacion

```python
from sklearn.metrics import classification_report
print(classification_report(y_prueba, y_pred))
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
confusion_matrix(y_prueba,y_pred)
```

```python
accuracy_score(y_prueba, y_pred)
```

<h2> Modelo K-NN <h2>

```python
from sklearn.neighbors import KNeighborsClassifier
```

<b> Partir en Train y Test <b>

```python
X_entreno, X_prueba, y_entreno, y_prueba = train_test_split(df.drop('Class' , axis = 1), df['Class'],
                                                    test_size = 0.30, random_state = 0)
```

<b>Encontrar valor K de menor error<b>

```python
errores = []
for i in range(1,50):
    knn = KNeighborsClassifier(n_neighbors = i)
    knn.fit(X_entreno, y_entreno)
    pred_i = knn.predict(X_prueba)
    errores.append(np.mean(pred_i != y_prueba))
```

```python
plt.figure(figsize = (10, 6))
plt.plot(range(1, 50), errores, color = 'blue', linestyle = 'dashed', marker = 'o',
         markerfacecolor = 'red', markersize = 10)
plt.title('Tasa de Error vs. Valor de K')
plt.xlabel('K')
plt.ylabel('Tasa de Error')
```

Encontramos que el menor valor de K para nuestro modelo seria de 3. Y se inputa al KNNeighbors Classifier

```python
# valor 3 vecinos
knn = KNeighborsClassifier(n_neighbors = 3)
knn.fit(X_entreno, y_entreno)
pred = knn.predict(X_prueba)
print(confusion_matrix(y_prueba,pred))
print(classification_report(y_prueba,pred))
```

# Hallazgos y conclusiones
Después de haber realizado los modelos de regresión logística y K-NN se observó que el rendimiento de dichos modelos tiene ciertas variaciones. Los valores de 'accuracy' obtenidos fueron de:
